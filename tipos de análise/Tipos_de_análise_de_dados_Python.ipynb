### Buscar padrões e tendências

```python
# Peso médio dos pinguins
df['body_mass_g'].mean()
```

```python
# Peso mínimo dos pinguins
df['body_mass_g'].min()
```

```python
# Peso máximo dos pinguins
df['body_mass_g'].max()
```

```python
# gráfico de quantidade de pinguins por espécie
sns.countplot(x=df['species'])
```

```python
# gráfico de quantidade de pinguins por ilha
sns.countplot(x=df['island'])
```

```python
# gráfico de quantidade de pinguins por ilha
sns.countplot(x=df['sex'])
```

```python
# gráfico de quantidade de pinguins por ilha e sexo
sns.countplot(x='island', hue='sex', data=df)
```

```python
# gráfico de quantidade de pinguins por espécie e sexo
sns.countplot(x='species', hue='sex', data=df)
```
