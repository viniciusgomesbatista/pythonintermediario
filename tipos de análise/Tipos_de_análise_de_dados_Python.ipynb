```python
# Criando subplots
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(12, 12), constrained_layout=True) #criando o subplot

# Gráficos de caixa para diferentes variáveis dos pinguins
sns.boxplot(x='species', y='bill_length_mm', hue='sex', data=df, orient='v', ax=axs[0][0])
sns.boxplot(x='species', y='bill_depth_mm', hue='sex', data=df, orient='v', ax=axs[0][1])
sns.boxplot(x='species', y='flipper_length_mm', hue='sex', data=df, orient='v', ax=axs[1][0])
sns.boxplot(x='species', y='body_mass_g', hue='sex', data=df, orient='v', ax=axs[1][1])

# Exibindo os subplots
plt.show()
```

```python
# gráficos de dispersão analisando tamanho do bico e altura do bico
sns.scatterplot(x='bill_length_mm', y='bill_depth_mm', data=df)
```

```python
# gráficos de dispersão analisando tamanho do bico e altura do bico por espécie
sns.scatterplot(x='bill_length_mm', y='bill_depth_mm', hue='species', data=df)
```

```python
# gráficos de dispersão analisando peso e tamanho da nadadeira
sns.scatterplot(x='body_mass_g', y='flipper_length_mm', data=df)
```

```python
# gráficos de dispersão analisando peso e tamanho da nadadeira por espécie
sns.scatterplot(x='body_mass_g', y='flipper_length_mm',hue='species', data=df)
```
