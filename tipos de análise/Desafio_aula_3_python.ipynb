```python
# Importar as bibliotecas
import pandas as pd # para trabalhar com dataframes
import seaborn as sns # para visualização de dados / gráficos
import matplotlib.pyplot as plt # para visualização de dados / gráficos
```

```python
# acessar os dados em um dataframe
df = sns.load_dataset(name='iris')
```

```python
df.head()
```

```python
df.columns
```

- sepal_length: comprimento da sépala
- sepal_width: largura da sépala
- petal_length: comprimento da pétala
- petal_width: largura da pétala
- species: espécie

```python
df.info()
```

```python
df.describe()
```

```python
# gráfico de peso médio dos pinguins por espécie
sns.barplot(x='species', y='sepal_length', data=df)
```

```python
# Visualizar a distribuição da massa corporal dos pinguis por espécie utilizando BoxPlot
plt.figure(figsize=(6,6))
sns.boxplot(y='sepal_length', data=df)
plt.show()
```

```python
sns.histplot(x='sepal_length', data=df, bins=100)  # Você pode ajustar o número de bins conforme necessário
plt.show()
```

```python
sns.violinplot(x='sepal_length', data=df)
plt.show()
```

```python
df.columns
```

```python
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(12, 12), constrained_layout=True)
sns.scatterplot(x='sepal_length', y='sepal_width', data=df, ax=axs[0][0])
sns.scatterplot(x='sepal_length', y='petal_length',  data=df, ax=axs[0][1])
sns.scatterplot(x='sepal_length', y='petal_width',  data=df, ax=axs[1][0])
```

```python
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(12, 12), constrained_layout=True)
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=df, ax=axs[0][0])
sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data=df, ax=axs[0][1])
sns.scatterplot(x='sepal_length', y='petal_width', hue='species', data=df, ax=axs[1][0])
```
